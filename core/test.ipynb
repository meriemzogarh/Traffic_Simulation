```python
import threading
import time

x=1

def function_to_run() -> int:
    for i in range(5):
        global x
        time.sleep(1)
        x+=1
        return x
        print("Hello from the thread")

# Create a thread
thread = threading.Thread(target=function_to_run)

# Start the thread
thread.start()

for i in range(10):
    print(thread)
    print(x)
    time.sleep(1)
# Wait for the thread to complete
thread.join()

print("Thread has finished execution")
```
